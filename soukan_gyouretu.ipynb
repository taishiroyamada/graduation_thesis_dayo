```python
#numpyのインストール
import numpy as np
#pandasのインストール
import pandas as pd
#statsmodelsのインストール
import statsmodels.formula.api as smf
#scikit-learnのインストール
import sklearn.linear_model as sl
#pyplotのインストール
import matplotlib.pyplot as plt
#seabornのインストール
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
w_data = pd.read_csv('data.csv')
w_data.shape
#195×22次元
```

```python
w_data.head(1)
```

```python
w_data = w_data.loc[:,['year','avg_temp','sun','avg_wind','rain_sum']]
#年月ごとのグラフ化
m_data = w_data.groupby('year').mean()
```

```python
m_data.plot()
```

```python
#カラム選択

#基本統計量の算出
w_data.describe()
```

```python
#データ確認
sns.pairplot(w_data)
```

```python
#相関行列
correlation_matrix = np.corrcoef(w_data.transpose())
w_data_name = w_data.columns.values
```

```python
#相関行列に名称をつける
sns.heatmap(correlation_matrix, annot=True, xticklabels=w_data_name,yticklabels=w_data_name)
```
